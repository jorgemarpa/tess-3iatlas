```python
import pandas as pd
import numpy as np 
from astroquery.jplhorizons import Horizons
from astropy.time import Time
from lxml import etree as XMLTree
from lxml.etree import Element as XMLElement
```

```python
lc = pd.read_csv("../data/3i_tess_s0092_time_series.csv", index_col=0)
lc
```

```python
obj = Horizons(id='-95', location='500', epochs=lc['time_utc'].values)
```

```python
obj.vectors(refplane='earth')
```

```python
mpc_ades = pd.DataFrame()
mpc_ades["permID"] = ["3I/ATLAS"] * 15
mpc_ades["trkSub"] = ["aa"] * 15
mpc_ades["stn"] = ["C57"] * 15
mpc_ades["mode"] = ["CCD"] * 15
mpc_ades["obsTime"] = Time(lc.time_utc.values, format="jd", scale="utc").isot
mpc_ades["ra"] = lc.ra.values
mpc_ades["dec"] = lc.dec.values
mpc_ades["rmsRA"] = [5] * 5 + [2] * 10
mpc_ades["rmsDec"] = [1.9] * 5 + [0.9] * 10
mpc_ades["mag"] = lc.prf_mag
mpc_ades["rmsMag"] = lc.prf_mag_err
mpc_ades["logSNR"] = np.log10(lc.prf_snr)
mpc_ades["band"] = ["T"] * 15
mpc_ades["fltr"] = ["T"] * 15
mpc_ades["exp"] = lc.time_window.values * 24 * 3600
mpc_ades["sys"] = ["ICRF_KM"] * 15
mpc_ades["ctr"] = 399
mpc_ades["pos1"] = obj.vectors(refplane='earth')["x"].to("km").value
mpc_ades["pos2"] = obj.vectors(refplane='earth')["y"].to("km").value
mpc_ades["pos3"] = obj.vectors(refplane='earth')["z"].to("km").value
mpc_ades["vel1"] = obj.vectors(refplane='earth')["vx"].to("km/s").value
mpc_ades["vel2"] = obj.vectors(refplane='earth')["vy"].to("km/s").value
mpc_ades["vel3"] = obj.vectors(refplane='earth')["vz"].to("km/s").value

mpc_ades
```

```python
utf8 = "UTF-8"

def createADESTree():
   """ createADESTree creates and returns
       a top-level ADES tree.  This returned
       pointer must be freed at the end
   """
   property = {}
   property["version"] = "2017"
   ades = XMLElement("ades", property)
   return ades
       

def addDataElement(branch, tag, text):
   """ addDataElement(branch, tag, text)
       makes a new element with tag and text
       extends branch with the new element (which now owns memory)
       and returns the new element for later extension itself
   """
   el = XMLElement(tag)
   el.text = text
   branch.extend([el])
   return el # so it can be extended itself

def addElement(branch, tag):
   """ addElement(branch, tag)
       makes a new element with tag
       extends branch with the new element (which now owns memory)
       and returns the new element for later extension itself
   """
   el = XMLElement(tag)
   branch.extend([el])
   return el # so it can be extended itself

def writeADESTree(filename, ades, xmlEncoding):
   """ writes the completed tree to filename
       with the requested encoding and returns
       the object to free.
   """
   mytree = XMLTree.ElementTree(ades)
   mytree.write(filename,
                pretty_print=True, 
                xml_declaration=True, 
                encoding=xmlEncoding)

def freeADESTree(tree):
   """ freeADESTree(tree) calls XMLFree on the
       input tree
   """
   pass # python does not need this since it collects garbage
```

```python
xmlEncoding = utf8

ades = createADESTree()

obsBlock = addElement(ades, "obsBlock")
obsContext = addElement(obsBlock, "obsContext")

item = addElement(obsContext, "observatory")
addDataElement(item, "mpcCode", "C57")
addDataElement(item, "name", "TESS")

item = addElement(obsContext, "submitter")
addDataElement(item, "name", "J. Martinez-Palomera")
addDataElement(item, "institution", "NASA Goddard Space Flight Center")

item = addElement(obsContext, "observers")
addDataElement(item, "name", "J. Martinez-Palomera")
addDataElement(item, "name", "A. Tuson")
addDataElement(item, "name", "C. Hedges")

item = addElement(obsContext, "measurers")
addDataElement(item, "name", "J. Martinez-Palomera")
addDataElement(item, "name", "A. Tuson")
addDataElement(item, "name", "C. Hedges")

item = addElement(obsContext, "telescope")
addDataElement(item, "aperture", "0.1")
addDataElement(item, "design", "Refractor")
addDataElement(item, "detector", "CCD")

item = addDataElement(obsContext, "fundingSource", "NASA")

item = addElement(obsContext, "comment");
addDataElement(item, "line", "TESS detections of 3I/ATLAS")
addDataElement(item, "line", "V to T from Woods et al 2021")

obsData = addElement(obsBlock, "obsData")

for k, row in mpc_ades.iterrows():
    optical = addElement(obsData, "optical")
    addDataElement(optical, "permID", "3I")
    addDataElement(optical, "trkSub", "aa")
    addDataElement(optical, "mode", "CCD")
    addDataElement(optical, "stn", "C57")
    addDataElement(optical, "obsTime", row["obsTime"][:-1]+"Z")
    addDataElement(optical, "ra", str(np.round(row["ra"], 6)))
    addDataElement(optical, "dec", str(np.round(row["dec"], 6)))
    # addDataElement(optical, "rmsRA", str(np.round(row["rmsRA"], 1)))
    # addDataElement(optical, "rmsDec", str(np.round(row["rmsDec"], 1)))
    # addDataElement(optical, "rmsCorr", "")
    addDataElement(optical, "astCat", "Gaia2")
    addDataElement(optical, "mag", str(np.round(row["mag"] + 0.671, 3)))
    addDataElement(optical, "rmsMag", str(np.round(row["rmsMag"], 3)))
    addDataElement(optical, "band", "V")
    addDataElement(optical, "logSNR", str(np.round(row["logSNR"], 3)))
    addDataElement(optical, "exp", str(np.round(row["exp"], 1)))
    
    
    
    freeADESTree(writeADESTree("../data/3i_tess_s0092_mpc.xml", ades, xmlEncoding))
```
